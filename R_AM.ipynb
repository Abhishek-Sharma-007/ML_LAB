```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder ,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.impute import SimpleImputer

df = pd.read_csv("module_5_auto.csv")
print("The first 5 rows of the dataframe") 
df.head(5)
```

```python
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df[['stroke','horsepower-binned']].describe
```

```python
imputer = SimpleImputer(missing_values = np.nan , strategy='mean')
df[['stroke']]= imputer.fit_transform(df[['stroke']])
```

```python
df['stroke'].isna().sum()
```

```python
df.dtypes
```

```python
df2 =df.apply(LabelEncoder().fit_transform)
df2.head(5)
```

```python
df2.isna().sum()
```

```python
df2.dtypes
```

```python
df2.corr()
```

```python
X = df2.iloc[: , 2:31]
y = df2.iloc[: , 26]
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size = 0.3, random_state = 0)
```

```python
print('X_train Shape:' , X_train.shape)
print('y_train Shape:' , y_train.shape)
```

```python
regressor = MLPRegressor(hidden_layer_sizes=(100,50), activation='relu', max_iter=2000 ,random_state=0 ,solver='lbfgs')
regressor
```

```python
regressor.fit(X_train , y_train)
```

```python
print(regressor.score(X_test , y_test)*100)
```

```python
da = pd.read_csv("module_5_auto.csv")
print("The first 5 rows of the dataframe") 
da.head(5)
```

```python
X = da[['horsepower','engine-size','peak-rpm']]
y = da['price']
```

```python
regressor = MLPRegressor(hidden_layer_sizes=(100,50), activation='relu', max_iter=5000 ,random_state=0 ,solver='lbfgs')
regressor
```

```python
regressor.fit(X_train , y_train)
```

```python
print(regressor.score(X_test , y_test)*100)
```
