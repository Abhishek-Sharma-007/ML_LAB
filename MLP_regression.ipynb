```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('housing.csv')
```

```python
data.describe()
```

```python
data.tail()
```

```python
data.isna().sum()
```

```python
#Handling null values by replacing it --->optional
smi = SimpleImputer(strategy='mean',missing_values=np.nan)
data['total_bedrooms'] = smi.fit_transform(data[['total_bedrooms']])
```

```python
data.isna().sum()
```

```python
#Handling null values by removing the rows ->>optional
data = data.dropna(axis='index',how='any')
```

```python
data.ocean_proximity.unique()
```

```python
data.corr()
```

```python
lbe = LabelEncoder()

data['ocean_proximity'] = lbe.fit_transform(data['ocean_proximity'])
```

```python
data
```

```python
# Drop Ocean_proximity column

data.drop('ocean_proximity',axis='columns', inplace=True )
```

```python
data
```

```python
X = data.iloc[:, 0:8]
y = data.median_house_value

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_std,y,test_size=0.2,random_state=2)
print('X Training Shape: ', X_train.shape)
print('Y Training Shape: ',y_train.shape)
print('X Test Shape: ', X_test.shape)
print('Y Test Shape: ',y_test.shape)
```

```python
mlp = MLPRegressor(hidden_layer_sizes = (200,140), activation = 'relu', solver = 'lbfgs')
mlp
```

```python
mlp.fit(X_train, y_train)
```

```python
print(mlp.score(X_train,y_train)*100)
```

```python
print(mlp.score(X_test,y_test)*100)
```
